```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
import os
import sys
import cv2
import numpy as np
import torch
import torchvision.transforms as transforms
from PIL import Image
from face_parsing.model import BiSeNet
from diffusers import StableDiffusionControlNetPipeline, ControlNetModel, UniPCMultistepScheduler
```

```python
def load_model(ckpt_path, n_classes=19):
    """
    Загружает модель BiSeNet с заданным количеством классов.
    """
    model = BiSeNet(n_classes)
    model.load_state_dict(torch.load(ckpt_path, map_location="cpu"))
    model.eval()
    return model
```

```python
def unify_labels(parsing_map):
    """
    Объединяет индексы, чтобы парные объекты имели один общий класс.
    Например, объединяет правый глаз и eye_g с левым глазом, правое ухо с левым и т.д.
    """
    parsing_map[parsing_map == 5] = 4   # правый глаз -> общий для глаз
    parsing_map[parsing_map == 6] = 4   # eye_g -> глаза
    parsing_map[parsing_map == 3] = 2   # объединить брови
    parsing_map[parsing_map == 8] = 7   # правое ухо -> общий для ушей
    parsing_map[parsing_map == 9] = 7   # ear_r -> уши
    parsing_map[parsing_map == 12] = 11 # u_lip -> губы
    parsing_map[parsing_map == 13] = 11 # l_lip -> губы
    parsing_map[parsing_map == 15] = 14 # neck_l -> шея
    return parsing_map
```

```python
def keep_largest_connected_component(parsing_map, label):
    """
    Для объекта, представленного одной областью, оставляет только самую крупную компоненту связности.
    """
    mask = (parsing_map == label).astype(np.uint8)
    num_labels, labels_cc, stats, _ = cv2.connectedComponentsWithStats(mask, connectivity=4)
    if num_labels > 1:
        largest_idx = 1 + np.argmax(stats[1:, cv2.CC_STAT_AREA])
        mask[...] = 0
        mask[labels_cc == largest_idx] = 1
    return mask
```

```python
def keep_two_largest_connected_components(parsing_map, label, min_area=50):
    """
    Для парного объекта сохраняет две крупнейшие компоненты связности, если их площадь не меньше min_area.
    """
    mask = (parsing_map == label).astype(np.uint8)
    num_labels, labels_cc, stats, _ = cv2.connectedComponentsWithStats(mask, connectivity=4)
    valid_components = []
    for i in range(1, num_labels):
        area = stats[i, cv2.CC_STAT_AREA]
        if area >= min_area:
            valid_components.append((i, area))
    valid_components = sorted(valid_components, key=lambda x: x[1], reverse=True)
    new_mask = np.zeros_like(mask)
    for comp in valid_components[:2]:
        comp_idx = comp[0]
        new_mask[labels_cc == comp_idx] = 1
    return new_mask
```

```python
def postprocess_connected_components(parsing_map, labels, paired_labels=set(), min_area=50):
    """
    Для каждого класса из списка labels:
      - Если класс является парным, оставляет две крупнейшие компоненты связности.
      - Иначе – оставляет только крупнейшую компоненту.
    Возвращает обновлённую карту сегментации.
    """
    new_map = parsing_map.copy()
    for lbl in labels:
        if lbl in paired_labels:
            cleaned_mask = keep_two_largest_connected_components(new_map, lbl, min_area)
        else:
            cleaned_mask = keep_largest_connected_component(new_map, lbl)
        new_map[new_map == lbl] = 0
        new_map[cleaned_mask == 1] = lbl
    return new_map
```

```python
def segment(im=str):
    # Параметры путей и классов
    ckpt_path = "79999_iter.pth"       # путь к весам модели
    img_path = im  # путь к исходному изображению
    n_classes = 19

    # 1. Загрузка модели
    model = load_model(ckpt_path, n_classes)

    # 2. Загрузка изображения и изменение его размера до 1024x1024
    image_pil = Image.open(img_path).convert("RGB")
    original_w, original_h = image_pil.size
    image_resized = image_pil.resize((1024, 1024), Image.BILINEAR)

    # 3. Преобразование изображения в тензор и запуск модели
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))
    ])
    img_tensor = transform(image_resized).unsqueeze(0)
    with torch.no_grad():
        output, _, _ = model(img_tensor)
    parsing_map = output.squeeze(0).argmax(0).cpu().numpy()

    # 4. Объединение парных классов
    parsing_map = unify_labels(parsing_map)

    # 5. Постобработка: оставляем только наибольшие компоненты связности
    labels_of_interest = [1, 4, 7, 10, 11, 14, 17]  # skin, eyes, ears, nose, mouth, neck, hair
    paired_labels = {4, 7, 11}  # глаза, уши, губы – парные объекты
    parsing_map = postprocess_connected_components(parsing_map, labels_of_interest, paired_labels, min_area=50)

    # 6. Приведение карты сегментации к исходному размеру и сохранение результатов
    parsing_map_full = cv2.resize(parsing_map.astype(np.uint8), (original_w, original_h), interpolation=cv2.INTER_NEAREST)
    np.save("segmentation_map.npy", parsing_map_full)
    print("Сегментационная карта успешно сохранена!")
```

```python
def get_face_geometry(skin_mask):
    """
    Анализирует форму лица по маске кожи.

    Возвращает:
      - face_type: качественное описание пропорций лица ("elongated", "broad", "balanced")
      - symmetry_desc: описание симметрии ("highly symmetrical" или "slightly asymmetrical")
      - face_rect: ограничивающий прямоугольник лица (x, y, w, h)
      - face_area: площадь лица (в пикселях)
      - geometry_description: подробное качественное описание формы лица
    """
    contours, _ = cv2.findContours(skin_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours:
        return None, None, None, None, "Face shape could not be determined."

    face_contour = max(contours, key=cv2.contourArea)
    x, y, w, h = cv2.boundingRect(face_contour)
    face_area = cv2.contourArea(face_contour)
    aspect_ratio = w / h

    # Качественная оценка пропорций лица
    if aspect_ratio < 0.9:
        face_type = "elongated"
    elif aspect_ratio > 1.1:
        face_type = "broad"
    else:
        face_type = "balanced"

    center_x = x + w // 2
    left_area = np.sum(skin_mask[:, :center_x]) / 255
    right_area = np.sum(skin_mask[:, center_x:]) / 255
    symmetry_ratio = min(left_area, right_area) / max(left_area, right_area) if max(left_area, right_area) > 0 else 1.0
    symmetry_desc = "highly symmetrical" if symmetry_ratio >= 0.9 else "slightly asymmetrical"

    # Формирование качественного описания формы лица без точных числовых значений
    if face_type == "elongated":
        shape_desc = "an elongated face that appears longer than it is wide"
    elif face_type == "broad":
        shape_desc = "a broad face with a wider horizontal dimension"
    else:
        shape_desc = "a well-balanced face with harmonious proportions"

    geometry_description = f"The subject has {shape_desc} and exhibits {symmetry_desc} features."
    return face_type, symmetry_desc, (x, y, w, h), face_area, geometry_description
```

```python
def analyze_hairline(hair_mask, face_rect):
    """
    Анализирует линию волос по маске волос.

    Возвращает качественное описание линии волос.
    """
    if face_rect is None:
        return "Hairline analysis unavailable."

    x, y, w, h = face_rect
    hair_contours, _ = cv2.findContours(hair_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if hair_contours:
        hair_contour = max(hair_contours, key=cv2.contourArea)
        hx, hy, hw, hh = cv2.boundingRect(hair_contour)
        rel_pos = (hy - y) / h  # относительное положение линии волос относительно лица
        if rel_pos < 0.15:
            hair_desc = "a very high hairline"
        elif rel_pos < 0.25:
            hair_desc = "a typical hairline"
        else:
            hair_desc = "a low hairline"
    else:
        hair_desc = "hairline is not clearly detected"
    return f"Hairline: {hair_desc}."
```

```python
def analyze_nose(nose_mask, face_area):
    """
    Анализирует размеры носа относительно площади лица.

    Возвращает качественное описание носа.
    """
    nose_area = np.sum(nose_mask) / 255
    ratio = nose_area / face_area if face_area > 0 else 0
    if ratio < 0.05:
        nose_desc = "a small nose"
    elif ratio < 0.1:
        nose_desc = "a moderate nose"
    else:
        nose_desc = "a prominent nose"
    return f"Nose: {nose_desc}."
```

```python
def analyze_ear(ear_mask, face_area):
    """
    Анализирует размеры ушей относительно площади лица.

    Возвращает качественное описание ушей.
    """
    ear_area = np.sum(ear_mask) / 255
    ratio = ear_area / face_area if face_area > 0 else 0
    if ratio < 0.03:
        ear_desc = "small ears"
    elif ratio < 0.07:
        ear_desc = "moderate ears"
    else:
        ear_desc = "large ears"
    return f"Ears: {ear_desc}."
```

```python
def generate_face_description_and_prompt(segmentation_map,
                                         skin_label=1,
                                         hair_label=17,
                                         nose_label=10,
                                         ear_label=7):
    """
    Создает бинарные маски для кожи, волос, носа и ушей, анализирует каждый параметр
    и формирует подробное текстовое описание, а также итоговый промт для генерации прически.
    """
    # Извлечение бинарных масок
    skin_mask = (segmentation_map == skin_label).astype(np.uint8) * 255
    hair_mask = (segmentation_map == hair_label).astype(np.uint8) * 255
    nose_mask = (segmentation_map == nose_label).astype(np.uint8) * 255
    ear_mask = (segmentation_map == ear_label).astype(np.uint8) * 255

    # Анализ формы лица
    face_type, symmetry_desc, face_rect, face_area, geom_desc = get_face_geometry(skin_mask)
    # Анализ линии волос
    hair_desc = analyze_hairline(hair_mask, face_rect)
    # Анализ носа и ушей
    nose_desc = analyze_nose(nose_mask, face_area)
    ear_desc = analyze_ear(ear_mask, face_area)

    # Сборка полного описания
    full_description = " ".join([geom_desc, hair_desc, nose_desc, ear_desc])

    # Рекомендации по прическе в зависимости от формы лица
    face_recommendations = {
        "elongated": "Consider a hairstyle that adds volume on the sides to balance the vertical elongation of the face.",
        "broad": "Consider a hairstyle that adds height and minimizes horizontal width, such as layered or updos.",
        "balanced": "A classic, well-proportioned hairstyle that maintains harmony with the natural features is ideal."
    }
    rec = face_recommendations.get(face_type, "")

    prompt = (f"{full_description} Generate a hairstyle that harmonizes with these features. "
              f"{rec} Ensure the style follows the natural hairline and enhances the facial symmetry.")

    return full_description, prompt
```

```python
def extract_hair_fragment_from_segmentation(original_img, segmentation_map, hair_label=17):
    """
    Извлекает фрагмент оригинального изображения, соответствующий области волос,
    используя карту сегментации (numpy array).

    Аргументы:
      original_img: исходное изображение (PIL Image).
      segmentation_map: карта сегментации (numpy array) с метками.
      hair_label: числовой идентификатор области волос (по умолчанию 17).

    Возвращает:
      hair_fragment: вырезанный фрагмент оригинального изображения, содержащий волосы.
    """
    # Создаем бинарную маску для волос из карты сегментации
    hair_mask_np = (segmentation_map == hair_label).astype(np.uint8) * 255
    hair_mask = Image.fromarray(hair_mask_np, mode="L")

    # Если размеры маски и оригинала не совпадают, изменяем размер маски
    if hair_mask.size != original_img.size:
        hair_mask = hair_mask.resize(original_img.size, resample=Image.NEAREST)

    # Используем маску для композиции: оставляем только области волос
    hair_fragment = Image.composite(original_img, Image.new("RGB", original_img.size, (0, 0, 0)), hair_mask)

    # Определяем ограничивающий прямоугольник по маске и обрезаем фрагмент
    bbox = hair_mask.getbbox()
    if bbox:
        hair_fragment = hair_fragment.crop(bbox)

    return hair_fragment
```

```python
def generate_hairstyle(prompt, conditional_image, num_inference_steps=30, guidance_scale=7.5):
    """
    Генерирует изображение с новой прической с помощью ControlNet.

    Аргументы:
      prompt: текстовый промпт для генерации прически.
      conditional_image: условное изображение (например, фрагмент волос, вырезанный из оригинала).
      num_inference_steps: количество шагов инференса (по умолчанию 30).
      guidance_scale: коэффициент управления (по умолчанию 7.5).

    Возвращает:
      сгенерированное изображение (PIL Image).
    """
    # Загружаем модель ControlNet (здесь используется модель sd-controlnet-hed для контроля контура)
    controlnet = ControlNetModel.from_pretrained("lllyasviel/sd-controlnet-hed", torch_dtype=torch.float16)

    # Загружаем pipeline Stable Diffusion с интеграцией ControlNet
    pipe = StableDiffusionControlNetPipeline.from_pretrained(
        "runwayml/stable-diffusion-v1-5",
        controlnet=controlnet,
        safety_checker=None,
        revision="main",
        torch_dtype=torch.float16
    )
    pipe.scheduler = UniPCMultistepScheduler.from_config(pipe.scheduler.config)
    pipe.enable_xformers_memory_efficient_attention()
    pipe.enable_model_cpu_offload()

    # Генерируем изображение
    output = pipe(prompt, conditional_image, num_inference_steps=num_inference_steps, guidance_scale=guidance_scale)
    return output.images[0]
```

```python
def diffuse_n_control():
    # Пример использования:
    # 1. Пусть оригинальное изображение хранится в файле "original.jpg"
    # 2. Карта сегментации сохранена в "segmentation_map.npy" (например, результат работы face_parsing)
    # 3. Промпт передается как аргумент командной строки или задается по умолчанию

    if len(sys.argv) > 1:
        prompt = sys.argv[1]
    else:
        prompt = ("A trendy modern hairstyle with intricate curls and natural color, "
                  "designed to enhance facial features and complement the hairline. "
                  "The style is vibrant, detailed, and harmonious.")

    original_image_path = "/petrov.jpg"
    segmentation_map_path = "/content/segmentation_map.npy"

    # Загружаем оригинальное изображение
    original_img = Image.open(original_image_path).convert("RGB")
    # Загружаем карту сегментации
    segmentation_map = np.load(segmentation_map_path)

    # Извлекаем фрагмент волос из оригинального изображения по карте сегментации
    hair_fragment = extract_hair_fragment_from_segmentation(original_img, segmentation_map, hair_label=17)

    # Генерируем изображение с новой прической, используя вырезанный фрагмент как условие
    generated_image = generate_hairstyle(prompt, hair_fragment, num_inference_steps=30, guidance_scale=7.5)

    # Сохраняем и отображаем результат
    generated_image.save("generated_hairstyle.png")
    generated_image.show()
```
